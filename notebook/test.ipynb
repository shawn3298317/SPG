```python
import os, sys
sys.path.append("../utils")

%load_ext autoreload
%autoreload 2

import matplotlib.patches as patches
import matplotlib.pyplot as plt
from LoadDataAiSky import *
```

```python
class Argument:
    batch_size = 1
args = Argument()
```

```python
_, _, tst_loader = data_loader(args)
```

```python
for i, (img, target) in enumerate(tst_loader):
    if i > 10: break
    #print(target[0]["bboxs"])
    
    if target[0]["label"] == 0: continue
    target_mask = np.zeros((224, 224))
    img = img[0].cpu().permute(1,2,0)
    plt.imshow(img)
    ax = plt.gca()
    for bbox in target[0]["bboxs"]:
        rect = patches.Rectangle((bbox.x1, bbox.y1), bbox.w, bbox.h, linewidth=2, edgecolor='lawngreen', facecolor='none', label="%.2f"%0.0)
        target_mask[int(bbox.y1):int(bbox.y2), int(bbox.x1):int(bbox.x2)] = 1.0
        ax.add_patch(rect)
    plt.show()
    
    plt.imshow(target_mask)
    plt.show()
```
